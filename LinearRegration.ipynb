```python
import pandas as pd
import seaborn as sns
```

```python
pg = sns.load_dataset('penguins')
pg.head(30)
```

Data Cleaning   


*   The reset_index() function resets the index values for the rows in the dataframe. Typically, you use reset_index() after you've finished manipulating the dataset. By setting inplace=True, you will not create a new DataFrame object. By setting drop=True, you will not insert a new index column into the DataFrame object.
* the dropna() function by default removes any rows with any missing values in any of the columns





```python
pg_sub = pg[pg['species']!='Chinstrap']
pg_final = pg_sub.dropna()
pg_final.reset_index(inplace = True, drop = True)
pg_final.head(30)
```

EDA

*  To visualize more than one relationship at the same time, we use the pairplot() function from the seaborn package to create a scatterplot matrix.



```python
sns.pairplot(pg_final)
```

Model Construction

```python
##To do this, you will first subset the variables of interest from the dataframe. You can do this by using double square brackets [[]], and listing the names of the columns of interest.
ol_data = pg_final[['bill_length_mm','body_mass_g']]
```

# formula

*  Next, you can construct the linear regression formula, and save it as a string. Remember that the y or dependent variable comes before the ~, and the x or independent variables comes after the ~.

Note: The names of the x and y variables have to exactly match the column names in the dataframe.



```python
ol_formula = 'body_mass_g ~ bill_length_mm'
```

```python
#Lastly, you can build the simple linear regression model in statsmodels using the ols() function. You can import the ols() function directly using the line of code below.
from statsmodels.formula.api import ols
```

Then, you can plug in the ols_formula and ols_data as arguments in the ols() function. After you save the results as a variable, you can call on the fit() function to actually fit the model to the data.

```python
OLS = ols(formula = ol_formula, data = ol_data)
model = OLS.fit()
```

```python
model.summary()
```

# You can use the regplot() function from seaborn to visualize the regression line.

```python
sns.regplot(data = ol_data, x = 'bill_length_mm',y = 'body_mass_g')
```

# Finish checking model assumptions¶
As you learned in previous videos, there are four main model assumptions for simple linear regression, in no particular order:

Linearity
Normality
Independent observations
Homoscedasticity

To more easily check the model assumptions and create relevant visualizations, you can first subset the X variable by isolating just the bill_length_mm column. Additionally, you can save the predicted values from the model using the model.predict(X) function.

```python
# Subset X variable
X = ol_data["bill_length_mm"]

# Get predictions from model
fitted_values = model.predict(X)
```

you can save the model residuals as a variable by using the model.resid attribute

```python
# Calculate residuals
residuals = model.resid
```

Check the normality assumption


*  Plot a histogram


```python
import matplotlib.pyplot as plt
fig = sns.histplot(residuals)
fig.set_xlabel("Residuals")
fig.set_title("Histogram")
plt.show()
```

```python
import statsmodels.api as sm
fig = sm.qqplot(residuals, line = 's')
plt.show()
```

Check the homoscedasticity assumption


```python
# Import matplotlib
import matplotlib.pyplot as plt
fig = sns.scatterplot(x=fitted_values, y=residuals)

# Add reference line at residuals = 0
fig.axhline(0)

# Set x-axis and y-axis labels
fig.set_xlabel("Fitted Values")
fig.set_ylabel("Residuals")

# Show the plot
plt.show()
```

# NEW EXAMPLES FOR SIMPLE LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
```

```python
data = pd.read_csv('/content/marketing_data.csv')
```

```python
data.head(10)
```

```python
data.isna().count()
```

```python
data.isna()
```

```python
data.shape
```

```python
data = data.dropna(axis=0)
```

```python
data.isna()
```

```python
sns.pairplot(data)
```

```python
#creating new dataFram
ol_data = data[['Radio','Sales']]
```

```python
ol_data.head()
```

```python
ol_formula = 'Sales ~ Radio'
```

```python
#applying Ordenary  Least Square for lr
OL = ols(formula = ol_formula, data = ol_data)
```

Create a linear regression model for the data and fit the model to the data.

```python
model = OL.fit()
```

Get a summary of the results from the model.

```python
model.summary()
```

Plot the OLS data with the best fit regression line.

```python
sns.regplot(data = ol_data, x = 'Radio', y = 'Sales')
```

###Check the normality assumption.
To get started, get the residuals from the model.

```python
residuals = model.resid
```

Visualize the residuals

```python
fig = sns.histplot(residuals)
fig.set_xlabel("Residuals")
fig.set_title("histogram")
plt.show()
```

Create a Q-Q plot to confirm the assumption of normality.

```python
sm.qqplot(residuals, line = 's')
plt.show()
```

```python
fit_value = model.predict(ol_data['Radio'])
```

```python
fig = sns.scatterplot(x = fit_value,y =residuals)
fig.axhline(0)
fig.set_xlabel("Fitted Values")
fig.set_ylabel("Residuals")
plt.show()
```
